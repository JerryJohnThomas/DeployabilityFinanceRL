```python
import numpy as np
import scipy
from sklearn.datasets import make_spd_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
n = 5
data_size = 5000
np.set_printoptions(precision=4)
np.random.seed(4) # keep 4
```

```python
plt.rcParams.update({
    #"text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": "Helvetica",
})
```

```python
cov = make_spd_matrix(n)/data_size/20
mu = np.random.randn(n)/data_size/40
# mu[0] = 0
# mu[n//2+1:n]  = -mu[n//2+1:n]
print(cov)
print(mu)
rdata = np.random.multivariate_normal(mu,cov,size=(data_size))
data = np.zeros((data_size+1,n)) + 5
for i in range(len(rdata)):
    data[i+1,:] = data[i,:]* np.exp(rdata[i,:])
```

```python
import seaborn as sns
sns.set_theme()
sns.set_style("whitegrid")
```

```python
fig, axs = plt.subplots(1,2,figsize=(15,5))
axs[1].plot(data,label=[i for i in range(1,n+1)])
axs[1].legend()
axs[0].plot(rdata,label=[i for i in range(1,n+1)])
axs[0].legend()
plt.show()
```
